```python
import pickle 
list_of_same_graphs = pickle.load(open('../data/list_of_same_graphs.pkl', 'rb'))
```

```python
import networkx as nx
import json 

for idx, list_of_graph in enumerate(list_of_same_graphs):
    graph = list_of_graph[0]
    nx.write_edgelist(graph['task_dag'].graph, f'../data/non_iso_graphs/graphs/{idx}.edgelist', data=False)

    pos = {}
    pos = graph['task_dag'].position 

    json.dump(pos, open(f'../data/non_iso_graphs/positions/{idx}.json','w'))

```

```python
import matplotlib.pyplot as plt 
graph_idx = 10
path = f'../data/non_iso_graphs/{graph_idx}.edgelist'

G_read = nx.read_edgelist(path, create_using=nx.DiGraph())
G_write = nx.DiGraph()
G_write.add_edges_from(list_of_same_graphs[graph_idx][0]['task_dag'].graph.edges)

pos = nx.spiral_layout(G_read)
nx.draw_networkx(G_read, pos=pos, with_labels=True)
print(G_read.edges)
plt.show()

pos = nx.circular_layout(G_write)
nx.draw_networkx(G_write, pos=pos, with_labels=True)
print(G_read.edges)
plt.show()
```
