```python
from utils.DAG_Generator import DAG

dag = DAG(5, 2)

import networkx as nx 
import matplotlib.pyplot as plt 

G = dag.graph

nx.draw(G,  pos=dag.position, with_labels=True, font_weight='bold', node_color='skyblue', edge_color='gray', node_size=800)

# plt.show()

import random

for (src, dest) in G.edges:
    G[src][dest]['demand'] = random.randint(1,10)

edge_labels = nx.get_edge_attributes(G, 'demand')
nx.draw_networkx_edge_labels(G, pos=dag.position, edge_labels=edge_labels)
plt.show()

# Processing Element Delay
for node in G.nodes:
    G.nodes[node]['delay'] = random.randint(11,20)

print("\n")
print(edge_labels)
print(nx.get_node_attributes(G, 'delay'))
```
